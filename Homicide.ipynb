```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# String que você quer buscar
string_para_buscar = 'All'

# Filtrando as linhas onde a string foi encontrada na coluna
filtro = df['Country'].astype(str).str.contains(string_para_buscar, case=False, na=False)
resultado = df.loc[filtro, 'Country'].unique()

# Mostrando o resultado
print(resultado)
```

```python
# Lista de nomes a serem removidos
nomes_para_remover = ['Asia', 'Europe', 'Americas', 'Africa', 'Oceania', 'World',
                      'Australia and New Zealand', 'Central Asia', 'Eastern Asia',
                      'Eastern Europe', 'Latin America and the Caribbean', 'Melanesia',
                      'Micronesia', 'Northern Africa', 'Northern America', 'Northern Europe',
                      'Polynesia', 'South-eastern Asia', 'Southern Asia', 'Southern Europe',
                      'Sub-Saharan Africa', 'Western Asia', 'Western Europe', 'All Africa',
                      'All Americas', 'All Asia', 'All Europe', 'All Oceania', 'World']

# Identificando as linhas a serem removidas
index_names = df[df['Country'].isin(nomes_para_remover)].index

# Removendo essas linhas
df = df.drop(index_names).reset_index(drop=True)
```

```python
# Visualizando os dados após a limpeza
df.head()
```

```python
# Visualizando os países disponivel no dataset após a limpeza
df['Country'].unique()
```

Quais países apresentam as 10 maiores índices de homicídios nos últimos 5 anos?

```python
# Filtrando os dados a partir de 2018 para um novo dataset
df_country = df.loc[df.Year >= 2018]

# Agrupando os dados por país
df_country = df_country[['Country', 'VALUE']]
df_country = df_country.groupby('Country').sum()

# Ordenando pelo valor em ordem decrescente
df_country = df_country.sort_values(by='VALUE', ascending=False).reset_index()
```

```python
# Visualizando os países com os 10 maiores índices de homicídios nos últimos 5 anos
df_country.head(10)
```

Quais países com menor número de homicídios em cada subregião?

```python
# Filtrando os dados por região em novos datasets
df_Americas = df.loc[df.Region == 'Americas']
df_Asia = df.loc[df.Region == 'Asia']
df_Europe = df.loc[df.Region == 'Europe']
df_Africa = df.loc[df.Region == 'Africa']
df_Oceania = df.loc[df.Region == 'Oceania']

# Agrupando os dados por país
df_Americas = df_Americas[['Country', 'VALUE']]
df_Americas = df_Americas.groupby('Country').sum()

df_Asia = df_Asia[['Country', 'VALUE']]
df_Asia = df_Asia.groupby('Country').sum()

df_Europe = df_Europe[['Country', 'VALUE']]
df_Europe = df_Europe.groupby('Country').sum()

df_Africa = df_Africa[['Country', 'VALUE']]
df_Africa = df_Africa.groupby('Country').sum()

df_Oceania = df_Oceania[['Country', 'VALUE']]
df_Oceania = df_Oceania.groupby('Country').sum()

# Ordenando pelo valor em ordem crescente
df_Americas = df_Americas.sort_values(by='VALUE').reset_index()
df_Asia = df_Asia.sort_values(by='VALUE').reset_index()
df_Europe = df_Europe.sort_values(by='VALUE').reset_index()
df_Africa = df_Africa.sort_values(by='VALUE').reset_index()
df_Oceania = df_Oceania.sort_values(by='VALUE').reset_index()
```

```python
# Criando coluna para adicionar as regiões no dataset
df_Americas['Region'] = 'Americas'
df_Asia['Region'] = 'Asia'
df_Europe['Region'] = 'Europe'
df_Africa['Region'] = 'Africa'
df_Oceania['Region'] = 'Oceania'
```

```python
# Armazendo os resultados em variáveis
Americas = df_Americas.head(1)
Asia = df_Asia.head(1)
Europe = df_Europe.head(1)
Africa = df_Africa.head(1)
Oceania = df_Oceania.head(1)
```

```python
# Concatenando as variáveis em um novo dataset
df_region_country = pd.concat([Americas, Asia, Europe, Africa, Oceania], axis=0)
```

```python
# Visualizando os países com  com menor número de homicídios em cada região
df_region_country
```

Qual a média de homicídios no Brasil nos últimos 10 anos?

```python
# Filtrando os dados de homicidios no Brasil a partir de 2013 para um novo dataset
df_Brasil = df.loc[df.Year >= 2013]
df_Brasil = df_Brasil.loc[df.Country == 'Brazil']

# Agrupando os dados por país
df_Brasil = df_Brasil[['Country', 'VALUE']]
df_Brasil = df_Brasil.groupby('Country').mean()

# Ordenando pelo valor em ordem decrescente
df_Brasil = df_Brasil.sort_values(by='VALUE', ascending=False).reset_index()
```
